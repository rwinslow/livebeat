```python
import time
import random
import datetime
import os

import pandas as pd
import numpy as np
import urllib3
```

```python
# Pull rechat messages for video_id=v84922619
# <Serenity> Gamescom Day 5, last stream here!

# The International 2016 - Main Event Day 4
# https://www.twitch.tv/dota2ti/v/83012529

# https://www.twitch.tv/dota2ti/v/83196893

time_string = "07:03:54"
t = datetime.datetime.strptime(time_string, "%H:%M:%S")
seconds = (60 * t.minute) + (3600 * t.hour) + t.second

index = t0 = 1470865301 # time 0
vid = 'v82878048'

try:
    print('Making', '/Volumes/Passport/LiveBeat/chat/{}'.format(vid))
    os.mkdir('/Volumes/Passport/LiveBeat/chat/{}'.format(vid))
except FileExistsError:
    pass
except:
    raise

http = urllib3.PoolManager()

print('Getting chats')
while index <= (t0 + seconds):
# while index <= 1471053328:
    url = 'rechat.twitch.tv/rechat-messages?start={}&video_id={}'.format(index, vid)
    output = '/Volumes/Passport/LiveBeat/chat/{}/rechat_{}_{}.json'.format(vid, vid, index)
    index += 30
    print(url)
    r = http.request('GET', url)
    with open(output, 'w') as out:
        out.write(r.data.decode('utf-8'))
    time.sleep(random.random()*3.5)
    
```
