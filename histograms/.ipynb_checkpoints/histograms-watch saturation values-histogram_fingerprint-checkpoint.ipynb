```python
import cv2
import numpy as np
from PIL import Image, ImageFilter
import imageio
import pandas as pd
from matplotlib import pyplot as plt
from scipy import spatial
from scipy.misc import imresize


%matplotlib inline
```

```python
def normalize(vec):
    _min = min(vec)
    _max = max(vec)
    return [(v - _min)/(_max - _min) for v in vec]

def get_histograms(video_path, seconds_between_frames=60, save=False, **kwargs):
    """Generate timecodes for game and non-game segments"""
        
    # Open file handle
    vid = imageio.get_reader(video_path, 'ffmpeg')

    # Get metadata and select 1 frame every n seconds
    meta = vid.get_meta_data()
    fps = int(np.round(meta['fps']))
    nframes = meta['nframes']
    
    try:
        start = kwargs['frame_range'][0]
        stop = kwargs['frame_range'][1]
        frames = np.arange(start, stop, seconds_between_frames*fps)
        print('Given frame range:', kwargs['frame_range'])
    except:
        print('No frame range. Using 100% of video')
        frames = np.arange(0, nframes, seconds_between_frames*fps)

    # Build histogram fingerprints
    histograms = []
    
    # Run through frames and find segments
    for i in frames:
        img = vid.get_data(i)

        # Resize image appropriately for training set
        img = imresize(img, (720, 1280))
        Image.fromarray(img).save(
            '/Volumes/Passport/LiveBeat/video/cosine/{}.png'.format(i)
        )
        histograms.append(Image.fromarray(img).histogram())

    # Close video handle to release thread and buffer
    vid.close()

    return histograms, frames

def compare_fingerprint(video_path, fingerprint, seconds_between_frames=60, **kwargs):        
    # Open file handle
    vid = imageio.get_reader(video_path, 'ffmpeg')

    # Get metadata and select 1 frame every n seconds
    meta = vid.get_meta_data()
    fps = int(np.round(meta['fps']))
    nframes = meta['nframes']
    
    frames = np.arange(0, nframes, int(seconds_between_frames*fps))

    highest_similarity = 0
    
    # Run through frames and find segments
    for i in frames:
        img = vid.get_data(i)

        # Resize image appropriately for training set
        img = imresize(img, (720, 1280))
        hist = normalize(Image.fromarray(img).histogram())
        similarity = get_similarity(fingerprint, hist)
        if similarity > highest_similarity:
            highest_similarity = similarity
            print('New highest similarity:', highest_similarity, '- frame', i)
            Image.fromarray(img).save('/Users/Rich/Desktop/highest.png')
        if  similarity > .99:
            vid.close()
            Image.fromarray(img).save(
                '/Users/Rich/Desktop/{}.png'.format(i)
            )
            return i

    # Close video handle to release thread and buffer
    vid.close()

    return histograms, frames


def get_similarity(vec1, vec2):
    return 1 - spatial.distance.cosine(vec1, vec2)
```
