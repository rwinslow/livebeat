```python
import cv2
import numpy as np
from PIL import Image, ImageFilter
import imageio
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def plotting(img):
    h, w, c = img.shape
    green_half = img[:,0:int(w/2),:]
    red_half = img[:,int(w/2):w,:]

    img = img[0:1,:,:]
    green_half_px = green_half[0:1,:,:]
    red_half_px = red_half[0:1,:,:]

    # Full image
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    h = cv2.calcHist([hsv], [0], None, [256], [0,256])
    s = cv2.calcHist([hsv], [1], None, [256], [0,256])
    v = cv2.calcHist([hsv], [2], None, [256], [0,256])
    h[0] = 0

    plt.figure()
    plt.plot(h)
    plt.plot(s)
    plt.plot(v)
    plt.title('Full image')
    plt.legend(['h', 's', 'v'])
    red_sum = int(sum(v[115:140]))
    green_sum = int(sum(v[85:105]))
    print('full variance', red_sum, green_sum)
    red_sum = int(sum(h[1:15]))
    green_sum = int(sum(h[55:65]))
    print('full hue', red_sum, green_sum)

    # Green half
    hsv = cv2.cvtColor(green_half_px,cv2.COLOR_BGR2HSV)
    h = cv2.calcHist([hsv], [0], None, [256], [0,256])
    s = cv2.calcHist([hsv], [1], None, [256], [0,256])
    v = cv2.calcHist([hsv], [2], None, [256], [0,256])
    h[0] = 0

    plt.figure()
    plt.plot(h)
    plt.plot(s)
    plt.plot(v)
    plt.title('Green half')
    plt.legend(['h', 's', 'v'])
    red_sum = int(sum(v[115:140]))
    green_sum = int(sum(v[85:105]))
    print('green variance', red_sum, green_sum)
    red_sum = int(sum(h[1:15]))
    green_sum = int(sum(h[55:65]))
    print('green hue', red_sum, green_sum)

    # Red half
    hsv = cv2.cvtColor(red_half_px,cv2.COLOR_BGR2HSV)
    h = cv2.calcHist([hsv], [0], None, [256], [0,256])
    s = cv2.calcHist([hsv], [1], None, [256], [0,256])
    v = cv2.calcHist([hsv], [2], None, [256], [0,256])
    h[0] = 0
    print('red sum:', sum(h[1:15]))

    plt.figure()
    plt.plot(h)
    plt.plot(s)
    plt.plot(v)
    plt.title('Red half')
    plt.legend(['h', 's', 'v'])
    red_sum = int(sum(v[115:140]))
    green_sum = int(sum(v[85:105]))
    print('red variance', red_sum, green_sum)
    red_sum = int(sum(h[1:15]))
    green_sum = int(sum(h[55:65]))
    print('red hue', red_sum, green_sum)
```

```python
plotting(cv2.imread('full.png'))
```

```python
img = cv2.imread('full_screen.png', cv2.IMREAD_UNCHANGED)
img = img[:,:,::-1] # OpenCV BGR -> RGB

h, w, c = img.shape

# Player status
factor = 0.226
y1 = int(h * 0.044)
x1 = int(w * (0.5 - factor))
y2 = y1+1
x2 = int(w * (0.5 + factor))
pl_status = img[y1:y2, x1:x2, :]

# Block out center of player status
h, w, c = pl_status.shape
factor = .12
x1 = int(np.round(w * (0.5 - factor)))+1
x2 = int(np.round(w * (0.5 + factor)))
pl_status[:, x1:x2, :] = 0

plt.figure(figsize=(16,2))
plt.imshow(pl_status)

img = pl_status[:,:,::-1] # Put player status from RGB to BGR for OpenCV
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
hist = cv2.calcHist([hsv], [0], None, [256], [0,256])
hist[0] = 0
plt.figure(figsize=(18,2))
plt.plot(hist)
plt.xticks(np.arange(0, 100,5))
plt.xlim((0, 100))

print(sum(hist[0:12]), sum(hist[55:65]))
```
